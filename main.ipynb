```python
from skvideo import io
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import ImageGrid
```

```python
raw_video = io.vread('WAAM-cut.mp4')
raw_video = raw_video[:,:,200:1200,:]
```

```python
fig = plt.figure(figsize=(10,10))
grid = ImageGrid(
    fig, 111, 
    nrows_ncols=(10, 10),
    axes_pad=0.1
)

for ax, im in zip(grid, raw_video[::10]):
    ax.imshow(im)
```
